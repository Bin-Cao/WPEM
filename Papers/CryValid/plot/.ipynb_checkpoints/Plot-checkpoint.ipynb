```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.font_manager import FontProperties
```

```python
int_99 = pd.read_csv('NTCAR99_Simu_profile.csv')
peak_99 = pd.read_csv('NTCAR99_Bragg_peaks.csv')['2theta/']

int_123 = pd.read_csv('TCAR123_Simu_profile.csv')
peak_123 = pd.read_csv('TCAR123_Bragg_peaks.csv')['2theta/']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use a style similar to Nature
#plt.style.use('seaborn-whitegrid')

# Define a color palette
colors = sns.color_palette("Set1")

font = {'family': 'Arial', 'weight': 'normal', 'size': 15}
font1 = {'family': 'Arial', 'weight': 'normal', 'size': 13}

fig, ax = plt.subplots(figsize=[7, 5])

# Plot profile#123 with the first color in the palette
ax.plot(int_123.iloc[:, 1], int_123.iloc[:, 2], color=colors[0], label="profile#123")
ax.vlines(x=peak_123, ymin=-9, ymax=-5, color=colors[0], linestyle='-', label="peaks#123")

# Plot profile#99 with the second color in the palette
ax.plot(int_99.iloc[:, 1], int_99.iloc[:, 2], color=colors[1], linestyle='--', label="profile#99")
ax.vlines(x=peak_99, ymin=-15, ymax=-11, color=colors[1], linestyle='-', label="peaks#99")

# Remove y-axis ticks
ax.yaxis.set_major_locator(plt.NullLocator())

plt.xlabel('2 ()', font)
plt.ylabel('I (a.u.)', font)

ax.legend(loc=1, prop=font1)

plt.tight_layout()  # Adjust layout for better spacing
plt.savefig('profile.png', dpi=800)
plt.savefig('profile.svg', dpi=800)
plt.show()
plt.clf()
```
