```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.font_manager import FontProperties
```

```python
expI = pd.read_csv('Eintensity.csv')
WPEMfit = pd.read_csv('WPEM.csv')


p1 = pd.read_csv('p1.csv')
peak_p1 = pd.read_csv('System0.csv')['mu_i']


p2 = pd.read_csv('p2.csv')
peak_p2 = pd.read_csv('System1.csv')['mu_i']

p3 = pd.read_csv('p3.csv')
peak_p3 = pd.read_csv('System2.csv')['mu_i']

p4 = pd.read_csv('p4.csv')
peak_p4 = pd.read_csv('System3.csv')['mu_i']
```

```python

font = {'family' : 'Arial',
'weight' : 'normal',
'size'   : 11,
}

font1 = {'family' : 'Arial',
'weight' : 'normal',
'size'   : 9,
}

fig, ax = plt.subplots(figsize=[8,5])

point_c = '#1f77b4'  # blue
point_s = 10

peak_c = '#d46f00'   # dark orange

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 12

ax.plot(p1.iloc[:,0], p1.iloc[:,1]+350, '-g',label= "decomposed : βSn")  # green
ax.vlines(x=peak_p1, ymin=320, ymax=340, color=peak_c, linestyle='-')

ax.plot(p2.iloc[:,0], p2.iloc[:,1]+250, '-b',label= "decomposed : Ag\u2083(Sn,In)")  # blue
ax.vlines(x=peak_p2, ymin=220, ymax=240, color=peak_c, linestyle='-')

ax.plot(p3.iloc[:,0], p3.iloc[:,1]+100, '-c',label= "decomposed : Cu\u2086(Sn,In)\u2085")  # cyan
ax.vlines(x=peak_p3, ymin=70, ymax=90, color=peak_c, linestyle='-')

ax.plot(p4.iloc[:,0], p4.iloc[:,1], '-m',label= "decomposed : Ti\u2082Sn\u2083")  # magenta
ax.vlines(x=peak_p4, ymin=-30, ymax=-10, color=peak_c, linestyle='-')

ax.scatter(expI.iloc[:,0], expI.iloc[:,1]+500, c=point_c, s=point_s ,label= "experiment")  # blue
ax.plot(WPEMfit.iloc[:,0], WPEMfit.iloc[:,1]+500, '-k', label= "WPEM fitting")

ax.yaxis.set_major_locator(plt.NullLocator())
plt.xlabel('2\u03b8\u00B0',font)
plt.ylabel('I (a.u.)',font)

ax.legend(loc=1,fontsize = 8)
plt.savefig('profile.png', dpi=800)
plt.savefig('profile.svg', dpi=800)
plt.show()
plt.clf()
```
