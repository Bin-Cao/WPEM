```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.font_manager import FontProperties
```

```python
int_pure = pd.read_csv('RuO2.csv')
fit_pure = pd.read_csv('WPEMRuO2.csv')
peak_pure = pd.read_csv('RuO2peaks.csv')['mu_i']

int_4 = pd.read_csv('4MI.csv')
fit_4 = pd.read_csv('WPEM4MI.csv')
peak_4 = pd.read_csv('4MIpeaks.csv')['mu_i']


int_8 = pd.read_csv('8MI.csv')
fit_8 = pd.read_csv('WPEM8MI.csv')
peak_8 = pd.read_csv('8MIpeaks.csv')['mu_i']

int_12 = pd.read_csv('12MI.csv')
fit_12 = pd.read_csv('WPEM12MI.csv')
peak_12 = pd.read_csv('12MIpeaks.csv')['mu_i']

Extinction_peak = pd.read_csv('RuO2_Extinction_peak.csv')['2theta/TOF']
```

```python

font = {'family' : 'Arial',
'weight' : 'normal',
'size'   : 11,
}

font1 = {'family' : 'Arial',
'weight' : 'normal',
'size'   : 9,
}


fig, ax = plt.subplots(figsize=[7,5])


point_c = 'r'
point_s = 15
peak_c = 'b'

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 12
    
    
ax.scatter(int_pure.iloc[:,0], int_pure.iloc[:,1], c=point_c, s=point_s ,label= "experiment", )
ax.plot(fit_pure.iloc[:,0], fit_pure.iloc[:,1], '-k', label= "WPEM", )
ax.vlines(x=peak_pure, ymin=-20000, ymax=-10000, color=peak_c, linestyle='-',label= "peaks")

ax.scatter(int_4.iloc[:,0], int_4.iloc[:,1]+100000, c=point_c, s=point_s ,)
ax.plot(fit_4.iloc[:,0], fit_4.iloc[:,1]+100000, '-k', )
ax.vlines(x=peak_4, ymin=80000, ymax=90000, color=peak_c, linestyle='-',)


ax.scatter(int_8.iloc[:,0], int_8.iloc[:,1]+200000, c=point_c, s=point_s ,)
ax.plot(fit_8.iloc[:,0], fit_8.iloc[:,1]+200000, '-k', )
ax.vlines(x=peak_8, ymin=180000, ymax=190000, color=peak_c, linestyle='-',)

ax.scatter(int_12.iloc[:,0], int_12.iloc[:,1]+300000, c=point_c,s=point_s , )
ax.plot(fit_12.iloc[:,0], fit_12.iloc[:,1]+300000, '-k', )
ax.vlines(x=peak_8, ymin=280000, ymax=290000, color=peak_c, linestyle='-',)

ax.axhline(y=-100000000,color='grey', linestyle='--',label= "extinction peaks",)
for k in range(len(Extinction_peak)):
    ax.axvline(x=Extinction_peak[k],color='grey', alpha=.3,linestyle='--',)

ax.yaxis.set_major_locator(plt.NullLocator())
ax.set_ylim(-50000, 410000)
plt.xlabel('2\u03b8\u00B0',font)
plt.ylabel('I (a.u.)',font)

ax.legend(loc=1,prop=font1)
plt.savefig('profile.png', dpi=800)
plt.savefig('profile.svg', dpi=800)
plt.show()
plt.clf()
```
